# Task
Find employees who earn more than their department's average salary from the `df` DataFrame and display their details.

## Calculate Department Averages

### Subtask:
Calculate the average salary for each department from the 'df' DataFrame and store it as a new DataFrame or Series.


**Reasoning**:
The subtask requires calculating the average salary for each department. This can be achieved by grouping the DataFrame by the 'department' column and then computing the mean of the 'salary' column. This result will be stored in a new variable called `department_avg_salaries`.



```python
department_avg_salaries = df.groupby('department')['salary'].mean()
print(department_avg_salaries)
```

## Merge with Employee Data

### Subtask:
Merge the calculated department average salaries back into the original 'df' DataFrame, ensuring each employee row has its department's average salary available for comparison.


**Reasoning**:
To ensure each employee row has its department's average salary, I will merge the `df` DataFrame with the `department_avg_salaries` Series on the 'department' column, and then update the `df` DataFrame.



```python
df = df.merge(department_avg_salaries.rename('department_average_salary'), left_on='department', right_index=True)
df
```

## Filter Employees

### Subtask:
Filter the merged DataFrame to select only those employees whose individual 'salary' is greater than their respective 'department_average_salary'.


**Reasoning**:
Filter the 'df' DataFrame to find employees whose salary is greater than their department's average salary, and store the result in a new DataFrame named 'employees_above_average'.



```python
employees_above_average = df[df['salary'] > df['department_average_salary']]
employees_above_average
```

## Display Results

Display the details of the employees who earn more than their department's average salary.
