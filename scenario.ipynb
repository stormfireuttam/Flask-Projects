```python
#Scenario 1: Employee Salary Analysis
import pandas as pd
data = [ [1,'IT',70000,6], [2,'HR',45000,4], [3,'IT',None,8], [4,'Finance',55000,3],
[5,'HR',38000,6] ]
df = pd.DataFrame(data, columns=['emp_id','department','salary','experience_years'])
df
```

```python
# 1. Which department has the highest average salary?
avg_salaries = df.groupby('department')['salary'].mean()
highest_dept = avg_salaries.idxmax()
highest_value = avg_salaries.max()

print(f"{highest_dept},{highest_value}")
```

```python
# 2. Which employees earn more than the department average?

df['dept_avg_salary'] = df.groupby('department')['salary'].transform('mean')
higher_earners = df[df['salary'] > df['dept_avg_salary']]
higher_earners
```

```python
# 3. Count employees with more than 5 years of experience per department
df_exp = df[df['experience_years'] > 5]
df_count = df_exp['department'].value_counts()

df_count
```

```python
#4.Which departments have minimum salary below 40,000?
df.groupby('department')['salary'].min().sort_values(ascending=True).head(1)
```

```python
#5.. Fill missing salary values using department median salary
df['salary'].fillna(df.groupby('department')['salary'].transform('median'), inplace=True)
df
```

```python
#Scenario 2: Orders Dataset
import pandas as ps
data = [ [101,1,2500,'2024-01-01'], [102,2,1800,'2024-01-02'],[103,1,None,'2024-01-02'], [104,3,3200,'2024-01-06'], [105,2,2100,'2024-01-07'] ]
df =ps.DataFrame(data, columns=['order_id','customer_id','order_amount','order_date'])
df['order_date']= ps.to_datetime(df['order_date'])
df

```

```python
#1. Which day had the highest total order value?
df_1=df.groupby('order_date')['order_amount'].max().idxmax()
df_1
```

```python
#2. Which customers placed more than 2 orders?
df.groupby('customer_id')['order_id'].count().loc[lambda x : x >=2]
```

```python
#3.What is the monthly total revenue?
df['month'] = df['order_date'].dt.to_period('M')
monthly_revenue = df.groupby('month')['order_amount'].sum()

monthly_revenue
```

```python
#4.Fill missing order amounts using daily average.
df['day']=df['order_date'].dt.day
```

```python
#5. Which orders were placed on weekends?
weekend_orders = df[df['order_date'].dt.dayofweek.isin([5, 6])]
weekend_orders
```

```python
#Scenario 3: Website Logs
import pandas as pd
data = [ [1,'/home',120,200], [2,'/login',300,500], [1,'/home',None,200],[3,'/dashboard',250,200], [3,'/dashboard',250,200] ]
df = pd.DataFrame(data,columns=['user_id','page','response_time','status_code'])
df
```

```python
#1. What is the average response time per page?
df.groupby('page')['response_time'].mean()
```

```python
#2. How many failed requests exist (status_code != 200)?
df.groupby('status_code')['user_id'].count().loc[lambda x : x != 200]
```

```python
#3.Which pages have response time above overall average
df['overall_avg'] = df['response_time'].mean()
df[df['response_time'] > df['overall_avg']]
```

```python
#4. Remove duplicate log entries.
df.duplicated()
```

```python
# 5. Fill missing response times using median.
df['response_time'] = df['response_time'].fillna(df['response_time'].median())
display(df)
```

```python
#Scenario 4: Student Scores
import pandas as pd
data = [ [1,'Math',78], [1,'Science',None], [2,'Math',35], [2,'Science',55],[3,'Math',90] ]
df = pd.DataFrame(data, columns=['student_id','subject','marks'])
df
```

```python
#1. What is the average marks per subject?
df_1=df.groupby('subject')['marks'].mean()
df_1
```

```python
#2. Which students scored below 40 in any subject?
df_1=df[df['marks'] < 40]
df_1
```

```python
#3. Count number of subjects per student.
df_1=df.groupby('student_id')['subject'].count()
df_1
```

```python
#4. Fill missing marks using subject average.
df['marks'] = df.groupby('subject')['marks'].transform(lambda x: x.fillna(x.mean()))
df
```

```python
#5. Which subject has the highest variance in marks?
df.groupby('subject')['marks'].var().idxmax()
```

```python
#Scenario 5: Product Inventory
import pandas as pd
data = [ [101,'Electronics',50,15000], [102,'Electronics',None,12000],[103,'Furniture',30,8000], [104,'Furniture',70,9000], [104,'Furniture',70,9000] ]
df = pd.DataFrame(data, columns=['product_id','category','stock','price'])
df
```

```python
#1. Which categories have total stock below 100?
df.groupby('category')['stock'].sum().loc[lambda x : x < 100]

```

```python
#2. Which products are priced above category average price?
df['category_avg_price'] = df.groupby('category')['price'].transform('mean')
df
```

```python
#3. Fill missing stock values using category median.
df['stock'] = df['stock'].fillna(df.groupby('category')['stock'].transform('median'))
df
```

```python
#4. Identify duplicate product records.
df.duplicated()
```

```python
# 5. What is the total inventory value per category?
df['inventory_value'] = df['stock'] * df['price']
total_inventory = df.groupby('category')['inventory_value'].sum()
display(total_inventory)
```
